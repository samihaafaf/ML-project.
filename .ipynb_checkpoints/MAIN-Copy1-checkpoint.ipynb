**LEBEL ENCODING FEATURES WITH BINARY CATAGORICAL VALUES + NOMINAL**

```python
from sklearn.preprocessing import LabelEncoder


categ = ['school',
 'sex',
 'address',
 'famsize',
 'Pstatus',
 'schoolsup',
 'famsup',
 'paid',
 'activities',
 'nursery',
 'higher',
 'internet',
 'romantic']

# Encode Categorical Columns 0/1
le = LabelEncoder()
new1 = new
new1[categ] = new1[categ].apply(le.fit_transform) #col = le.fit_trnasform(col)
new1

#the features in categ are transfomed into binary values
```

```python
#checking shape
new1.shape
```

```python
#pd.get_dummies() using one hot encoding increases the colomun by 13 which is a lot, so we use frequency encoding
lis = ['Mjob','Fjob','reason','guardian']
pd.get_dummies(new,columns = lis).shape
```

```python
#checking the categories of the cols
lis = ['Mjob','Fjob','reason','guardian']
for each in lis:
    print(each, len(new1[each].unique()),'lebels')
```

```python
#checking count for each category
for each in lis:
    print(each,new1[each].value_counts().to_dict())
#here no two category in a colomn has the same frequency so we may do freq encoding
```

```python
#using count/frequency encoding for Mjob, Fjob, reason, guardian.
feq_map_Fjob = new1['Fjob'].value_counts().to_dict()
feq_map_Mjob = new1['Mjob'].value_counts().to_dict()
feq_map_reason = new1['reason'].value_counts().to_dict()
feq_map_guardian = new1['guardian'].value_counts().to_dict()
```

**Frequency Count Encoding**

```python
new2 = new1.copy() #new dataframe to keep new1 intact
new2['Fjob'] = new2['Fjob'].map(feq_map_Fjob)
new2['Mjob'] = new2['Mjob'].map(feq_map_Mjob)
new2['reason'] = new2['reason'].map(feq_map_reason)
new2['guardian'] = new2['guardian'].map(feq_map_guardian)
new2

```

```python
new2.corr()
```

```python
new3 = new2.copy()
new3 = new3.drop(['famsize','Pstatus','Mjob','Fjob','reason','famsup','paid','activities','nursery','romantic','famrel','health'],axis = 1)





new3.shape
```

```python
x = new3.iloc[:,:19].values
y = new3.iloc[:,19].values
```

**KNN**

```python
#y1 = pd.DataFrame(y)
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=73
                                                   )
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
print("Test set predictions:\n", y_pred)
```

```python
count = 0
for i in range( len(y_test) ):
    if y_pred[i] != y_test[i]: 
        count = count + 1
error = count/len(y_pred)
print( "Error = %f " % (error*100) + '%' )
accuracy = (1-error)
print( "Accuracy = %f " % (accuracy*100) + '%' ) #33% after deleting low correlation cols
```

**LOGISTIC REGRESSION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=73)
```

```python
#running logistic
model = LogisticRegression()
model.fit(x_train, y_train) #Training the model
predictions = model.predict(x_test)
print(predictions)# printing predictions
```

```python
print( accuracy_score(y_test, predictions))   #30% after deleting low corr cols
```
